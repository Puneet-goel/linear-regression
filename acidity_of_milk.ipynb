```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
url ="https://raw.githubusercontent.com/Puneet-goel/learn/main/linearX.csv"
X = pd.read_csv(url)

url="https://raw.githubusercontent.com/Puneet-goel/learn/main/linearY.csv"
Y = pd.read_csv(url)

X=(X.values).reshape((99,))
Y=(Y.values).reshape((99,))

print(X)
print(Y)
```

```
plt.scatter(X,Y)
plt.show()
```

```
X = X-X.mean()/(X.std())

plt.scatter(X,Y)
plt.show()
```

```
def function(X,theta):
   return theta[0]+theta[1]*X


def error(X,Y,theta):
  ans = 0.0
  n =X.shape[0]
  for i in range(n):
    ans += (Y[i]-function(X[i],theta))**2

  return ans*0.5


def gradient(X,Y,theta):
   grad=np.array([0.0,0,0])
   n = X.shape[0]

   for i in range(n):
      grad[0]-= (Y[i]-function(X[i],theta)) 
      grad[1]-= (Y[i]-function(X[i],theta))*X[i]

   return grad

 
def gradient_descent(X,Y,learning_rate , iterations) :
    theta=np.array([0.0,0.0])
    #print(learning_rate)
    err = []
    for i in range(iterations):
       grad = gradient(X,Y,theta)
       a = error(X,Y,theta)
       theta[0] = theta[0]-learning_rate*grad[0]
       theta[1] = theta[1]-learning_rate*grad[1]
       err.append(a)


    return theta,err

```

```
theta , e = gradient_descent(X,Y,learning_rate=0.001,iterations=500)
print(theta)
print(e)
```

```
plt.plot(e)
plt.show()
```

```
plt.scatter(X,Y)
plt.plot(X,function(X,theta),color='y')
plt.show()
```
